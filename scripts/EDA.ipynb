```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pylab as plt
df = pd.read_csv('../data/data.csv')
print(df.dtypes) 
# object = string
```

```python

# Check for zero values in 'price' column
num_zero = np.sum(df['price'] == 0)
print(f"Number of zero prices: {num_zero}")

plt.figure(figsize=(8,5))
plt.hist(np.log1p(df['price']), bins=200, color="#029DBC",edgecolor='black', linewidth=0.5)
plt.xlabel('log(Price + 1) [USD]', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of Real Estate Prices (Log-Transformed)', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

plt.show()

#

```

```python

# Remove NA
df_valid = df[(df['price'] >= 0) & (df['house_size'] > 0)].dropna(subset=['price', 'house_size'])

nbins = 60  

heatmap, xedges, yedges = np.histogram2d(
    np.log1p(df_valid['price']), 
    np.log(df_valid['house_size']),
    bins=nbins
)

# avoide log(0)
heatmap[heatmap == 0] = 0.1

extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

plt.figure(figsize=(6,4))
plt.imshow(np.log10(heatmap).T, origin='lower', extent=extent, cmap='viridis', vmin=0)
plt.xlabel('log(Price+1) [USD]')
plt.ylabel('log(House Size) [sqrt]')
plt.title('Distribution Density of House Size and Price (log-transformed)', fontsize=10, fontweight='bold')
plt.colorbar(label='log10(count)')
plt.tight_layout()
plt.show()
```
